```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# creating engine to hawaii.sqlite
engine = create_engine("sqlite:///../Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
print(Base.classes.keys())
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
Last_date=session.query(Measurement.date).order_by(Measurement.date.desc()).first()
print(Last_date)
```

```python
# Designing a query to retrieve the last 12 months of precipitation data and plot the results. 
a_year_ago = dt.date(2017, 8, 23) - dt.timedelta(days=365)
precip_data = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= a_year_ago).all()

# Save the query results as a Pandas DataFrame
precip_df = pd.DataFrame(precip_data,columns=['Date', 'precipitation'])

# Sort the DataFrame by date and printing the first 5 colummns
raw_precip_df = precip_df.sort_values('Date')
print(raw_precip_df.head())  

```

```python
#cleanning the data since there are some NaN values
df_cleaned = precip_df.dropna()
```

```python
# Using Pandas Plotting with Matplotlib to plot the data
df_cleaned.plot(x='Date', y='precipitation', rot=90)
plt.ylabel('Inches')
plt.show()
```

```python
# Summary statistics for the precipitation data
df_cleaned.describe()
```

# Exploratory Station Analysis

```python
# Query to calculate the total number of stations in the dataset
total_stations = session.query(func.count(Station.station)).all()
print(total_stations)
```

```python

# query to find the most active stations (i.e. which stations have the most rows?) and Listing the stations and their counts in descending order.
active_stations = session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).all()

# Print the results
active_stations
```

```python
#finding the most active station
most_active_station=active_stations[0][0]
most_active_station
```

```python
# Using the most active station id from the previous query, calculating the lowest, highest, and average temperature.
temp_values = session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
filter(Measurement.station == most_active_station).all()
print(temp_values [0])
```

```python
# Using the ,Quering the last 12 months of temperature observation data for the most active station
tobs_values = session.query(Measurement.date, Measurement.tobs).filter(Measurement.station == most_active_station) \
                    .filter(Measurement.date >= a_year_ago).all()
tobs_df = pd.DataFrame(tobs_values, columns=["date", "tobs"])# Creating a dataframe of the query in order to plot it
```

```python
# Plotting the temperature observations as a histogram
tobs_df.plot(kind="hist")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```

# Close Session

```python
# Close Session
session.close()
```
